```python
#SIMPLE LINEAR REGRESSION

#importing required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#loading the dataset
df=pd.read_csv("student_scores - student_scores.csv")
print(df.head())
print(df.shape)
```

```python
#defining the dependent and independent variables
x=df[['Hours']]
y=df['Scores']
```

```python
#splitting the dataset into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#training the model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#printing the intercept and coefficient
print("Intercept:",model.intercept_)
print("Coefficient:",model.coef_)
```

```python
#evaluating the model
from sklearn.metrics import mean_squared_error,r2_score
y_pred=model.predict(x_test)
print("mse:", mean_squared_error(y_test,y_pred))
print("r2_score:", r2_score(y_test,y_pred))
```

```python
#visualizing the results
import matplotlib.pyplot as plt
plt.scatter(x_test,y_test, label="Actual scores")
plt.plot(x_test,y_pred,label="Regression Line")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Hours vs Scores")
plt.legend()
plt.show()

```

```python
#predicting the score for 9.25 hours of study
predicted_score=model.predict([[9.25]])
print("Predicted score for 9.25 hours of study:",predicted_score[0])
```
