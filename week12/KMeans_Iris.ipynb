```python
#demonstatration of kmeans clustering algorithm - iris dataset

#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
#load dataset
df=load_iris()
x=df.data
y=df.target
print(x.shape)
```

```python
#standardize
sc=StandardScaler()
x_sc=sc.fit_transform(x)
```

```python
#apply kmeans
km=KMeans(n_clusters=3,random_state=42,n_init=10)
km.fit(x_sc)
```

```python
#get cluster labels
cl=km.labels_
print("cluster labels:",np.unique(cl))
```

```python
#cluster centers
print("cluster centers:", km.cluster_centers_)
```

```python
#visualize clusters
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
xpca=pca.fit_transform(x_sc)
plt.figure(figsize=(6,4))
plt.scatter(xpca[:,0],xpca[:,1],c=cl, cmap='viridis')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('K-Means Clustering')
plt.show()
```
