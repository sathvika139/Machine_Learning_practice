```python
 #Demonstration of PCA- principal component Analysis - iris dataset

 #importing required libraries
 import numpy as np
 import matplotlib.pyplot as plt
 from sklearn.datasets import load_iris
 from sklearn.decomposition import PCA
 from sklearn.preprocessing import StandardScaler
```

```python
#load dataset:
df=load_iris()
x=df.data
y=df.target
print("data shape:", x.shape)
print("target shape:", y.shape)
```

```python
#Standardize the dataset-pca is scale-dependent

sc=StandardScaler()
x_std=sc.fit_transform(x)
```

```python
#apply pca
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_std)
print("reduced data shape:", x_pca.shape)
```

```python
#explained variance ratio

print("explained variance ratio:", pca.explained_variance_ratio_)
print("total variance explained:", np.sum(pca.explained_variance_ratio_))
```

```python
#visualize the pca result

plt.figure(figsize=(6,4))
plt.scatter(x_pca[y==0,0], x_pca[y==0,1], label='Malignant', alpha=0.7)
plt.scatter(x_pca[y==1,0], x_pca[y==1,1], label='Benign', alpha=0.7)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Result')
plt.legend()
```
