```python
#Numpy is a powerful library for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
matrix = np.array([[1, 2], [3, 4]])

print(f"Mean: {np.mean(arr)}")
print(f"Matrix Transpose:\n{matrix.T}")
```

```python
#Pandas is a versatile library for data manipulation and analysis. It provides data structures like Series and DataFrame, which allow for easy handling of structured data, making it ideal for tasks such as data cleaning, transformation, and analysis.
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['NY', 'LA', 'CHI']
}
df = pd.DataFrame(data)

print(df[df['Age'] > 28])
```

```python
#Matplotlib is a widely used library for creating static, animated, and interactive visualizations in Python. It provides a flexible and powerful interface for generating a wide variety of plots and charts, making it an essential tool for data visualization.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o', color='green', linestyle='--')
plt.title("Simple Line Trend")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

```python
#Seaborn is a statistical data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics, making it easier to explore and understand complex datasets through visualizations.
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load a built-in dataset
df = sns.load_dataset("iris")

# 2. Create a scatter plot with one line of code
# It automatically handles the colors (hue) and legend
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")

plt.show()
```

```python
#Scikit-learn is a powerful machine learning library in Python that provides simple and efficient tools for data mining and data analysis. It offers a wide range of algorithms for classification, regression, clustering, and dimensionality reduction, making it a popular choice for building machine learning models.
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data: X (Square footage), y (Price)
X = np.array([[1000], [1500], [2000], [2500]])
y = np.array([300000, 450000, 590000, 720000])

# Initialize and train the model
model = LinearRegression()
model.fit(X, y)

# Predict price for a 1800 sq ft house
prediction = model.predict([[1800]])
print(f"Predicted Price: ${prediction[0]:,.2f}")
```
